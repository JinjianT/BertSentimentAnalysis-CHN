```python
import os
import json
import pandas as pd
import csv
import math
```

```python
emotion_dict = {"中性":0,"正面":1,"负面":2}
```

```python
with open("XiEMO1.csv","w",newline = "",encoding="utf-8") as csvfile:
    fieldNames = ['emotion','text','content']
    writer = csv.DictWriter(csvfile, fieldnames=fieldNames)
    writer.writeheader()
    for line in open("XiEMO1.txt", encoding="utf-8"):
        data = json.loads(line)
        data['emotion'] = emotion_dict[data['emotion']]
        writer.writerow(data)
```
